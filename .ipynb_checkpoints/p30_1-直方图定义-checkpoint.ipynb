## **直方图**

![](./image/P30_1.png)

**cv2.calcHist(images,channels,mask,histSize,ranges)**

* images：原图像图像格式为uint8或float32。当传入函数时应用中括号[]括来，例如[img]
* channels：同样用中括号括来。如果图像是灰度图它的值就是[0]，如果是彩色图像则传入参数可以是[0][1][2]，它们分别对应着BGR。
* mask：掩模图像。如果是整幅图像的直方图则为None。如果是某一部分的直方图，则需要传入一个掩膜。
* histSize：BIN的数目。也应用中括号[]括住。
* ranges：像素值范围为[0,256]

```python
img = cv2.imread('./image/P4_2_cat.jpg',0) #0表示灰度图
hist = cv2.calcHist([img],[0],None,[256],[0,256])
hist.shape
```

```python
plt.hist(img.ravel(),256)
plt.show()
```

```python
img = cv2.imread('./image/P4_2_cat.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
```
